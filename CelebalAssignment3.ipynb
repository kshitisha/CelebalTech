```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("revenue_target.csv")
```

```python
#here i am checking the revenue file data that i have uploaded
print(df.head())
print(df.info())


print(df.describe())

#setting style accordingly
sns.set(style="whitegrid")
```

```python
#this is visualization 1 - distribution of Target Revenue
plt.figure(figsize=(10,6))
sns.histplot(df["Target Revenue"], kde=True, color='green')
plt.title("Distribution of Target Revenue")
plt.xlabel("Target Revenue")
plt.ylabel("Frequency")
plt.show()
```

```python
#this is visualization 2 - average Revenue per Quarter
plt.figure(figsize=(8,6))
sns.barplot(x="quarter", y="Target Revenue", data=df, estimator='mean', ci=None, palette="muted")
plt.title("Average Target Revenue per Quarter")
plt.ylabel("Average Revenue")
plt.show()
```

```python
#this is visualization 3 - Revenue by Card Category
plt.figure(figsize=(10,6))
sns.boxplot(x="Card_Category", y="Target Revenue", data=df, palette="coolwarm")
plt.title("Target Revenue by Card Category")
plt.show()
```

```python
#this is visualization 4 - Revenue Trend by Year
plt.figure(figsize=(10,6))
sns.lineplot(x="Year", y="Target Revenue", data=df, estimator='mean', ci=None)
plt.title("Revenue Trend Over Years")
plt.ylabel("Average Target Revenue")
plt.show()
```

```python
#this is visualization 5 - Type of Purchase vs Revenue
plt.figure(figsize=(8,6))
sns.violinplot(x="Type", y="Target Revenue", data=df)
plt.title("Revenue Distribution by Purchase Type")
plt.show()
```
